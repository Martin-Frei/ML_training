# Linear Regression


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#  Create dataset (Pizza size vs. Price)
data = {'Diameter': [8, 10, 12], 'Price': [10, 13, 16]}
df = pd.DataFrame(data)

# Split into X (feature) and y (target)
X = df[['Diameter']]
y = df['Price']
```

```python
# Create Linear Regression Model
model = LinearRegression()

# Train (Fit) the model on data
model.fit(X, y)
```

```python
# Predict price for new pizza sizes (e.g., 14 inches)
new_sizes = np.array([[14], [16], [20]])
predictions = model.predict(new_sizes)

print("Predicted Prices:", predictions)
```

```python
# Plot Data Points
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot Regression Line
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')

# Labels and Title
plt.xlabel("Pizza Diameter (inches)")
plt.ylabel("Price ($)")
plt.title("Pizza Size vs. Price")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predict on training data
y_pred = model.predict(X)

# Calculate MSE and R² Score
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create dataset (Pizza size vs. Price)
data = {'study_hours': [2, 4, 6], 'score': [17, 29, 46]}
df = pd.DataFrame(data)

# Split into X (feature) and y (target)
X = df[['study_hours']]
y = df['score']
```

```python
# Create Linear Regression Model
model = LinearRegression()

# Train (Fit) the model on data
model.fit(X, y)
```

```python
# Predict price for new pizza sizes (e.g., 14 inches)
new_records = np.array([[1], [5], [7], [9]])
predictions = model.predict(new_records)

print("Predicted new_records:", predictions)
```

```python
# Plot Data Points
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot Regression Line
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')

# Labels and Title
plt.xlabel("study_hours")
plt.ylabel("scores")
plt.title("study_hours' vs. score")
plt.legend()
plt.show()
```

```python
print("Slope (m):", model.coef_[0])
print("Y-Intercept (b):", model.intercept_)
```
