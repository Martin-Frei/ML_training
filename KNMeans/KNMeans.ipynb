```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score

# # Sample Dataset
# data = np.array([
# [15, 40], [16, 42], [17, 39], [18, 80], [19, 90],
# [20, 88], [25, 20], [30, 30], [33, 85], [35, 90],
# [45, 45], [46, 50], [50, 60], [55, 65], [60, 70],
# [65, 75], [70, 15], [75, 20], [80, 25], [85, 30]
# ])

# # Visualizing Data
# plt.scatter(data[:, 0], data[:, 1], s=100, color='gray')
# plt.title("Customer Data (Before Clustering)")
# plt.xlabel("Annual Income (in $1000s)")
# plt.ylabel("Spending Score (0-100)")
# plt.show()

# Synthetic dataset: Age vs. Subscription Duration (months)
data = np.array([
    # Young users with short subscriptions (Cluster 1)
    [18, 2], [20, 3], [22, 1], [19, 2], [21, 3], [23, 1], [25, 2], [24, 1],
    
    # Middle-aged users with medium subscriptions (Cluster 2)
    [35, 12], [40, 15], [38, 10], [42, 14], [45, 16], [50, 18], [48, 12], [52, 15],
    
    # Older users with long subscriptions (Cluster 3)
    [60, 24], [65, 30], [70, 36], [68, 28], [72, 32], [75, 34], [80, 40], [78, 38],
    
    # Some noise/outliers
    [30, 25], [55, 5], [28, 20], [62, 10]
])

# Visualize raw data
plt.scatter(data[:, 0], data[:, 1], s=80, color='blue', alpha=0.7)
plt.title("User Data: Age vs. Subscription Duration")
plt.xlabel("Age (years)")
plt.ylabel("Subscription Duration (months)")
plt.grid(True)
plt.show()

# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualizing Results
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=100)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("User Data: Age vs. Subscription Duration with K-Means")
plt.legend()
plt.show()

# Evaluation
print("Inertia (Lower is Better):", kmeans.inertia_)
print("Silhouette Score (Higher is Better):", silhouette_score(data, labels))

# Display Results
print("Cluster Labels:", labels)
print("Centroids:\n", centroids)


# Some noise/outliers  [30, 25], [55, 5], [28, 20], [62, 10]
```
